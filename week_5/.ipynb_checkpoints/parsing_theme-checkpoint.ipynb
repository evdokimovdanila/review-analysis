```python
import requests as req
import bs4
```

Выделение основыных заголовков

```python
respons_1 = req.get("https://en.wikipedia.org/wiki/Bias-variance_tradeoff")
print(respons_1.status_code)
```

```python
print(respons_1.text)
respons_1.encoding = 'utf-8'
```

Чтобы выделить основыне заголовки сделаем фильтр по тегам h1 и h2. При этом в тегом 2 также размечены: Сontents и Navigation menu, если нужны только содержательные заголовки, то можно сделать дополнительную фильтрацию

```python
parser = bs4.BeautifulSoup(respons_1.text, 'lxml')
heading = parser.findAll(["h1","h2"])
for i in heading:
    print(i.text)
```

Только содержательные заголовки

```python
parser = bs4.BeautifulSoup(respons_1.text, 'lxml')
list_h = [] 

heading1 = parser.find("h1")
list_h.append(heading1.text)

heading2 = parser.findAll("h2")
for h2 in heading2:
    gen_h2 = h2.find("span", attrs={"class":"mw-headline"})
    if gen_h2 != None:
        list_h.append(gen_h2.text)
print('\n'.join(list_h))
```

Темы статей по машинному обучению

```python
response2 = req.get("https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms")
response2.status_code
```

```python
response2.text
```

```python
parser2 = bs4.BeautifulSoup(response2.text, 'lxml')
names = parser2.find("div", attrs={"id":"mw-pages"})# находим обертку под ML.
names_topic = names.findAll("li")# выделяем названия статей.
for name in names_topic:
    print(name.text)
```
